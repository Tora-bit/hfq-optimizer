```python
import numpy as np
def nominalic(L,ic, log=False):
    phizero=2.07e-15
    x = L * ic /phizero
    if log:
        print("L:\t",L,"[H]")
        print("Ic:\t",ic,"[A]")
        print("LIc:\t",L * ic,"[Wb]")
        print("LIc/Φ0:\t",L * ic / phizero)
    return ( 0.5263*x**6 - 2.3279*x**5 + 3.4434*x**4 - 1.0023*x**3 - 2.5876*x**2 + 3.341*x ) * ic

def Lic(icn, log=False):
    if icn < 0 or 1.5 < icn:
        raise ValueError("Out of range value")

    phizero=2.07e-15
    rs_list = np.roots([0.5263, -2.3279, 3.4434, -1.0023, -2.5876, 3.341, -icn])
    ans =  [ i.real for i in rs_list if i.imag == 0 and i.real >= 0]
    lic_phizero = ans[0]
    if log:
        print("Icn:\t",icn,"[A]")
        print("LIc/Φ0:\t",lic_phizero)
    return lic_phizero * phizero
```

```python
Lic(50e-6, log=True)
```

```python
import matplotlib.pyplot as plt
x = [ i*1e-6 for i in range(100) ]
y = []
for i in x:
    y.append(Lic(i)/4e-12)

print(y)
plt.figure(figsize=(10,7))
plt.tick_params(labelsize=15)
# 緑色のサインカーブを描く
plt.plot(x, y, color = "green")
```
